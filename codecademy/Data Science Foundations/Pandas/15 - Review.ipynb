```python
# Import necessary libraries
import pandas as pd

# Task 1 - Load the CSV file into a DataFrame
orders = pd.read_csv('./_resources/Pandas - Shoefly.csv')
print(orders)

# Task 2 - Display the first 5 rows of the DataFrame
print(orders.head())

# Task 3 - Extract the 'email' column and print it
emails = orders['email']
print(emails)

# Task 4 - Filter the DataFrame for orders made by Frances Palmer
frances_palmer = orders[orders.first_name == "Frances"]
print(frances_palmer)

# Task 5 - Filter the dataframe by shoe type in the list ["clogs", "boots", "ballet flats"]
comfy_shoes = orders[orders.shoe_type.isin(["clogs", "boots", "ballet flats"])]
print(comfy_shoes)
```
